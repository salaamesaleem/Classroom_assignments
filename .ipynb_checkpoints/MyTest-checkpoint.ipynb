```python


# Load your CSV
df = pd.read_csv(r"List.csv") # Using 'df' for the DataFrame name as good practice

# Create the boolean Series for each condition
# These are Series of True/False, NOT DataFrames
condition_vcf = df['narrative1'].str.contains("vcf", case=False, na=False)
condition_gold = df['narrative1'].str.contains("gold", case=False, na=False)

# Combine the boolean Series using the pandas | (OR) operator
# This 'comb' will be a single boolean Series
comb = condition_vcf | condition_gold

# Use this combined boolean Series to filter the original DataFrame
df_filtered = df[comb]

#print(df_filtered)
```

```python

import numpy as np
import pandas as pd
df = pd.read_csv(r"List.csv")
condition_vcf = df['narrative1'].str.contains("vcf",case=False,na=False)
condition_gold = df['narrative1'].str.contains("gold",case=False,na=False)
condition_agr = df['narrative1'].str.contains("agr",case=False,na=False)
#comb = condition_vcf | condition_gold | condition_agr

condition = [condition_vcf,condition_gold,condition_agr]
choice = ['vcf','gold','agr']
defvalue ='other'
df['catagory']=np.select(condition , choice , default = defvalue)

#df_filtered = df[comb]
#print(df_filtered)
print(df)
```
